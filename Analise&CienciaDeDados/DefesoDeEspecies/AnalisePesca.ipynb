### Duração dos Períodos de Defeso

Quais são os períodos de defeso mais longos e mais curtos?


```python
print(PescaDF["INICIO"].unique())
```

```python
# Verificar o número de linhas válidas nas colunas 'INICIO' e 'TÉRMINO'
validos_inico = PescaDF['INICIO'].notna().sum()
validos_termino = PescaDF['TÉRMINO'].notna().sum()

# Verificar a quantidade de valores válidos em cada coluna
print(f"Linhas válidas para 'INICIO': {validos_inico}")
print(f"Linhas válidas para 'TÉRMINO': {validos_termino}")

# Contar o total de linhas válidas para a análise de duração
validos_total = PescaDF.dropna(subset=["INICIO", "TÉRMINO"]).shape[0]
print(f"Linhas válidas para análise de duração: {validos_total}")
```

```python
# Corrigindo meses com espaçoes extras e substituição de textos invalidos
def limpar_mes(data):
    meses_map = {
        "janeiro": "01",
        "fevereiro": "02",
        "março": "03",
        "abril": "04",
        "maio": "05",
        "junho": "06",
        "julho": "07",
        "agosto": "08",
        "setembro": "09",
        "outubro": "10",
        "novembro": "11",
        "dezembro": "12",
    }

    data = str(data).strip().lower()

    # Substituir meses por extenso
    for mes,num in meses_map.items():
        data = data.replace(mes, num)

    # Removendo qualquer valor não relacionado a uma data válida
    if "ano" in data or "I" in data or "II" in data or "III" in data:
        return None  # Valor inválido, pode retornar None (NaT) ou NaN
    return data

# Aplicando a limpeza nas colunas de data
PescaDF["INICIO"] = PescaDF["INICIO"].apply(limpar_mes)
PescaDF["TÉRMINO"] = PescaDF["TÉRMINO"].apply(limpar_mes)


```

```python
# Converter para datetime, ignorando valores inválidos
PescaDF["INICIO"] = pd.to_datetime(PescaDF["INICIO"], errors='coerce', dayfirst=True)
PescaDF["TÉRMINO"] = pd.to_datetime(PescaDF["TÉRMINO"], errors='coerce', dayfirst=True)
```

```python
# Excluir linhas com datas inválidas (NaT)
PescaDF = PescaDF.dropna(subset=["INICIO", "TÉRMINO"])

```

```python
# Verificar o resultado
print(PescaDF.head())
```

```python
# Calcular a duração entre 'TÉRMINO' e 'INICIO'
PescaDF['DURACAO'] = (PescaDF['TÉRMINO'] - PescaDF['INICIO']).dt.days

# Gerar estatísticas sobre a duração
estatisticas_duracao = PescaDF['DURACAO'].describe()
print(estatisticas_duracao)
```

```python
# Plotar o histograma
plt.figure(figsize=(10, 6))
plt.hist(PescaDF['DURACAO'].dropna(), bins=15, color='skyblue', edgecolor='black')

# Adicionar título e rótulos
plt.title('Distribuição da Duração dos Períodos de Defeso', fontsize=14)
plt.xlabel('Duração (dias)', fontsize=12)
plt.ylabel('Frequência', fontsize=12)

# Exibir o gráfico
plt.grid(True)
plt.show()


```
