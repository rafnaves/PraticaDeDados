## Analise Bivariada

```python
# Hipotese: Clientes com contrato do tipo mensal s√£o mais propensos ao churn
# Construir tabela de contigencia entre tipo de contrato e churn

pd.crosstab(df_churn.Churn, df_churn.Contract, margins=True)
```

```python
# Contruir Tabela de Contigencia entre Tipo de Contrato e Churn - propor√ß√£o
pd.crosstab(df_churn.Churn, df_churn.Contract, margins=True, normalize='index')
```

Considerando o publico total, os clientes de contrato mensal representam 55%. porem se consideramos apenas os clientes que abandonaram o servi√ßo, 88% tinham contrato mensal mes.
Desta forma, podemos afirmar que h√° uma correla√ß√£o entre i tipo de contrato mensal e de abandono

```python
# avaliar correla√ß√£o entre duas variaveis categoricas

# Teste de hipotese chamado chi-square ou qui-quadrado de pearson
# num teste de hipotese, duas hipoteses s√£o formuladas:
# H0 (hipotese nula): as duas variaveis s√£o idependentes
# h1 (hipotese complementar): as duas n√£o s√£o idependentes

# o teste confirma ou recusa a hipotese nula
#  quando a probabilidade de observarmos h0 √© inferior a 0.05 (p-value) recusamos a hipotese nula

```

```python
# df da crosstab (sem totais)
df_crosstab_churn_contract = pd.crosstab(df_churn.Churn, df_churn.Contract)
```

```python
# Calcular os scores e p-values
from scipy.stats import chi2
from scipy.stats import chi2_contingency
```

```python
# Tupla = Lista imutavel
chi_scores_churn_contract = chi2_contingency(df_crosstab_churn_contract)
```

```python
chi_scores_churn_contract
```

```python
scores_churn_contract = pd.Series(chi_scores_churn_contract[0])
```

```python
pvalues_churn_contract = pd.Series(chi_scores_churn_contract[1])
```

```python
# apresentar numeros com decimais sem a nota√ß√£o cientifica
pd.set_option('display.float_format', lambda x: '%.15f' %x)
```

```python

# Criar dataframe com os resultados
df_chi_scores_churn_contract = pd.DataFrame({'qui2': scores_churn_contract, 'p_value': pvalues_churn_contract})
```

```python
df_chi_scores_churn_contract
```

Conforme Teste estatistico (chi square), o pvalue <= 0.05, desta forma rejeitamos a hipotese nula, ou seja as variaveis n√£o s√£o idependentes. pelo qui2 alto, podemos afirmar que h√° uma forte correla√ß√£o

```python
# Avaliar a correla√ß√£o entre uma variaval qualitativa q quantitativa
# Usar o mesmo teste de chi-square, mas transformar a variavel quantitativa em qualitativa
# hipotese cliente com menos de 6 meses de contrato √© mais propenso ao churn

# criar uma coluna nova no DataFrame com base numa condi√ß√£o de outra coluna
# coluna = 'TempoMenor6Meses'
import numpy as np
```

```python
# criar coluna nova
df_churn['TempoMenor6Meses'] = np.where(df_churn.tenure<6, 'Yes', 'no')
```

```python
df_churn.head(5)
```

```python

df_crosstab_churn_tenure = pd.crosstab(df_churn.Churn, df_churn.TempoMenor6Meses)
```

```python
#calcular os scores
chi_scores_churn_tenure = chi2_contingency(df_crosstab_churn_tenure)
```

```python
scores_churn_tenure = pd.Series(chi_scores_churn_tenure[0])
pvalues_churn_tenure = pd.Series(chi_scores_churn_tenure[1])
```

```python
df_chi_scores_churn_tenure = pd.DataFrame({'Qui2': scores_churn_tenure, 'p-value':pvalues_churn_tenure})
df_chi_scores_churn_tenure
```

Como o p-value <= 0.5, rejeitamos a h0, ou seja as variaveis n√£o s√£o idependentes
- a correla√ß√£o entre churn e tempo de contrato <6 meses √© menor que a correla√ß√£o entre churn e tipo de contrato.

Isso significa que h√° uma associa√ß√£o estatisticamente significativa entre idade acima de 65 anos e churn. Em outras palavras, a idade do cliente influencia o churn de forma relevante.

```python
df_churn_month = df_churn[df_churn.Contract =='Month-to-month']
df_churn_month
```

```python
df_churn_month.TotalCharges.plot.hist()
```

```python
# Metodo de Tukey - IQR (DISTRIBUI√á√ÉO N√ÉO NORMMAL)]
# IQR - Range Interquartil
# IQR = 3o quartil - 1o Quartil
q1_TotalCharges_month = df_churn_month.TotalCharges.quantile(0.25)
q3_TotalCharges_month = df_churn_month.TotalCharges.quantile(0.75)
iqr_TotalCharges_month = q3_TotalCharges_month - q1_TotalCharges_month
iqr_TotalCharges_month
```

```python
limInf_TotalCharges_month = q1_TotalCharges_month - (iqr_TotalCharges_month * 1.5)
limSup_TotalCharges_month = q3_TotalCharges_month + (iqr_TotalCharges_month * 1.5)
limInf_TotalCharges_month, limSup_TotalCharges_month
```

```python
df_churn_month[(df_churn_month.TotalCharges < limInf_TotalCharges_month) |
               (df_churn_month.TotalCharges > limInf_TotalCharges_month) ]
```

```python
# ZScore (usado para distribui√ß√£o normal)
# Indica quantos desvios padr√µes um ponto especifico de dados est√° distante da m√©dia
# zscore = (x - m√©dia) / desvio padr√£o
from scipy.stats import zscore
```

```python
z = np.abs(zscore(df_churn_month.TotalCharges))
```

```python
z
```

```python
df_churn_month[z > 3.0]
```
