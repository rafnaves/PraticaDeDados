## Dicionario

````python
import pandas as pd
import matplotlib.pyplot as plt

# Dicionário de faturamento
dict_faturamento = {
    'data_ref': [
        '2023-01-01', 
        '2020-02-01', 
        '2021-03-01', 
        '2022-04-01', 
        '2023-05-01',
        '2023-06-01', 
        '2020-07-01', 
        '2021-08-01', 
        '2022-09-01', 
        '2023-10-01',
        '2022-11-01', 
        '2023-12-01',
        ],
    'valor': [
        400000, 
        890000, 
        760000, 
        430000, 
        920000,
        340000, 
        800000, 
        500000, 
        200000, 
        900000,
        570000, 
        995000,
        ]
}
````

### Tarefas
Trazer a média das vendas

Criar um gráfico de linhas mostrando o mês de referência e o valor.

Criar um gráfico de barras HORIZONTAL mostrando o mês de referência e o valor

````python
df_faturamento = pd.DataFrame.from_dict(dict_faturamento)
df_faturamento
````

````python
## Media das vendas
df_faturamento["valor"].mean()
````

Média: 642.083,33

````python
## Criar um gráfico de linhas mostrando o mês de referência e o valor.
df_faturamento['data_ref'] = pd.to_datetime(df_faturamento['data_ref'])
df_faturamento.plot.line(x='data_ref', y='valor')
````

### Por que usar isso?
Os valores no formato de texto não permitem cálculos diretos ou manipulações úteis, enquanto os valores em formato **datetime** permitem que você trabalhe facilmente com datas e horários.

### Exemplo prático:
Antes da conversão:
```plaintext
data_ref
0    "2023-01-01"
1    "2020-02-01"
```

Depois da conversão:
```plaintext
data_ref
0    2023-01-01 00:00:00
1    2020-02-01 00:00:00
```

````python
## Criar um gráfico de barras HORIZONTAL mostrando o mês de referência e o valor.
df_faturamento.sort_values('valor', ascending=True).plot.barh(x='data_ref', y='valor')
````
